```python
posi = data_df.dropna(subset=["Latitude"])
```

```python
posi.isnull().sum().sort_values(ascending=False)
```

```python
def convert_currency(value):
    # 这里value是经度在前，纬度在后
    return value.split(',')[0]

posi['mylat'] = posi['Latitude'].apply(convert_currency)
```

```python
def convert_currency(value):
    return value.split(',')[1]

posi['mylon'] = posi['Latitude'].apply(convert_currency)
```

```python
posi
```

```python
def convert_unitPrice(value):
    """
    转换字符串数字为float类型
     - 移除单价,  /平米,
     - 转化为float类型
    """
    new_value = value.replace('单价', '').replace('元/平米', '')
    return np.float(new_value)

posi['average_price'] = posi['average_price'].apply(convert_unitPrice)
posi
```

```python
lat = np.array(posi["mylat"][0:len(posi)])
lon = np.array(posi["mylon"][0:len(posi)])
```

```python
posi
```

```python
price = np.array(posi["average_price"][0:len(posi)],dtype=float)
data1 = [[lon[i],lat[i],price[i]] for i in range(len(posi))]
data1
```
