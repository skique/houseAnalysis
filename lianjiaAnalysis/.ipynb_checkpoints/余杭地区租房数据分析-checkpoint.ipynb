```python
# 面积
def convert_currency(value):
    value = value.split('㎡')[0]
    new_value = value.replace('--', '')
    return new_value

df['area'] = data_df['houseInfo'].apply(convert_currency)
```

```python
# 朝向
def convert_currency(value):
    return ''.join(re.findall('㎡(.+)\d室',value))
     
df['direction'] = data_df['houseInfo'].apply(convert_currency)
df
```

```python
# 户型
def convert_currency(value):
    return ''.join(re.findall('\d室\d厅\d卫',value))
     
df['layout'] = data_df['houseInfo'].apply(convert_currency)
df
```

```python
# 经度和纬度
def convert_currency(value, num):
    res = re.findall('\((.+)\)',value)
    
    if len(res):
        return res[0].split(',')[num]
    else:
        return np.nan
    
     
df['lng'] = data_df['position'].apply(convert_currency, args=(1,))
df['lat'] = data_df['position'].apply(convert_currency, args=(0,))

df
```
