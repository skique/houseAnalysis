```python
#重新排列特征位置
columns = ['title','house','totalPrice','unitPrice','area','houseInfo']
df = pd.DataFrame(data_df, columns = columns)
```

```python
df
```

```python
def convert_currency(value):
    """
    转换字符串数字为float类型
     - 移除 ￥ ,
     - 转化为float类型
    """
    new_value = value.replace('万', '0000')
    return np.float(new_value)

df['totalPrice'] = df['totalPrice'].apply(convert_currency)
```

```python
def convert_unitPrice(value):
    """
    转换字符串数字为float类型
     - 移除单价,  /平米,
     - 转化为float类型
    """
    new_value = value.replace('单价', '').replace('元/平米', '')
    return np.float(new_value)

df['unitPrice'] = df['unitPrice'].apply(convert_unitPrice)
df
```

数据处理，去除不需要的字段，将总价和单价处理成数值型，之后要对这两个字段进行处理

## size特征

```python
# 先从houseInfo字段中取到面积字段，再转成数值形
def convert_size(value, num):
    """
    转换字符串数字为float类型
     - 移除 ￥ ,
     - 转化为float类型
    """
    size = value.split('|')
    new_value = size[num].replace('平米', '') if num == 1 else size[num]
    
    return np.float(new_value) if num == 1 else new_value

df['layout'] = df['houseInfo'].apply(convert_size, args=(0,))
df['size'] = df['houseInfo'].apply(convert_size, args=(1,))
df['direction'] = df['houseInfo'].apply(convert_size, args=(2,))
df['renovation'] = df['houseInfo'].apply(convert_size, args=(3,))
df
```

```python
f,[ax1, ax2]=plt.subplots(1,2,figsize=(15,8))

#房屋面积分布情况
sns.distplot(df['size'],bins=20,kde=True,ax=ax1,hist_kws={'alpha':1,'color':'r'})
ax1.set_title('房屋面积分布', fontsize=15)
#房屋面积和出售价格关系
sns.regplot(x='size', y='totalPrice', data=df, ax=ax2)
ax2.set_title('房屋面积和出售价格关系', fontsize=15)
plt.show()
```

(1)建造面积大部分在150平方以内，超过150平方米的二手房多为别墅或者复试结构。

(2)根据散点图可发现，Size和totalPrice呈现正相关，房屋面积越大，房价越高，和符合实际情况

```python
#Direction特征
def direct_func(x):
    if not isinstance(x,str):
        raise TypeError
    x = x.strip()
    x_len = len(x)
    x_list = pd.unique([y for y in x])
    if x_len != len(x_list):
        return ''.join(y)
        
    if (x_len == 2) & (x not in d_list_two):
        m0 = x[0]
        m1 = x[1]
        return m1+m0
    elif (x_len == 3) & (x not in d_list_three):
        for n in d_list_three:
            if (x_list[0] in n) & (x_list[1] in n) & (x_list[2] in n):
                return n
    elif (x_len == 4) & (x not in d_list_four):
        return d_list_four[0]
    else:
        return x
    
# 对“Direction”特征
d_list_one = ['东','西','南','北']
d_list_two = ['东西','东南','东北','西南','西北','南北']
d_list_three = ['东西南','东西北','东南北','西南北']
d_list_four = ['东西南北']    
df['Direction'] = df['direction'].apply(direct_func)
```
