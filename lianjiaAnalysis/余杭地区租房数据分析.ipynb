```python
# 经度和纬度
def convert_currency(value, num):
    res = re.findall('\((.+)\)',value)
    
    if len(res):
        return res[0].split(',')[num]
    else:
        return np.nan
    
     
df['lng'] = data_df['position'].apply(convert_currency, args=(1,))
df['lat'] = data_df['position'].apply(convert_currency, args=(0,))

df
```

```python
# 整租或者合租
def convert_currency(value):
    if len(value):
        return value.split('·')[0]
    else:
        return np.nan
    
     
df['methods'] = data_df['title'].apply(convert_currency)
df
```

```python
# 房型-室
def convert_currency(value):
    str = ''.join(re.findall('(\d)室',value))
    return pd.to_numeric(str)
    
     
df['room_num'] = data_df['houseInfo'].apply(convert_currency)
df
```

```python
# 房型-厅
def convert_currency(value):
    str = ''.join(re.findall('(\d)厅',value))
    return pd.to_numeric(str)
    
     
df['living_num'] = data_df['houseInfo'].apply(convert_currency)
df
```

```python
bins = list(range(500, 10000, 500))
df['priceRange'] = pd.cut(df['price'], bins)
df
```

```python
# 对租金进行分组后对比租房数量
df_house_count = df.groupby('priceRange')['price'].count().sort_values(ascending=False).to_frame().reset_index()
df_house_count
```

```python
# 对区域进行分组后对比房租均价
df_house_price = df.groupby('city')['price'].mean().sort_values(ascending=False).to_frame().reset_index()
df_house_price
```

```python
import seaborn as sns
```

```python
plt.style.use("bmh")
plt.rc('font',family='SimHei', size=13)
%matplotlib inline

f,[ax1,ax2] = plt.subplots(2,1,figsize=(15,24))

sns.barplot(x='priceRange', y='price', palette="Greens_d", data=df_house_count, ax=ax1)
ax1.set_title('房租价格分布',fontsize=20)
ax1.set_xlabel('价格')
ax1.set_ylabel('数量')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)

sns.barplot(x='city', y='price',palette="Blues_d", data=df_house_price, ax=ax2)
ax2.set_title('各区房租均价分布',fontsize=20)
ax2.set_xlabel('区域')
ax2.set_ylabel('均价')

plt.show()
```
